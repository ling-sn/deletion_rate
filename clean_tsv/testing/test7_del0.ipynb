```python
from pathlib import Path
import traceback
import pandas as pd
import numpy as np
import re
from scipy.stats import fisher_exact
```

### Loading in data

```python
current_path = Path.cwd()
input_dir = current_path/"test7_del0_output"
nuc = pd.read_csv(input_dir/"Nuc.tsv", sep = "\t")
cyto = pd.read_csv(input_dir/"Cyto.tsv", sep = "\t")
```

### Analyzing data

```python
## Drop duplicate GenomicModBases to obtain rows with unique genes
nuc_unique = nuc.drop_duplicates(subset = ["GenomicModBase"], keep = "first")
cyto_unique = cyto.drop_duplicates(subset = ["GenomicModBase"], keep = "first")

print(f"Length of Nuc: {len(nuc_unique)}",
      f"Length of Cyto: {len(cyto_unique)}", sep = "\n")
```

```python
from IPython.display import display, HTML, Image
dfs = [nuc_unique, cyto_unique]
names = ["nuc", "cyto"]
for df, name in zip(dfs, names):
    df = df.copy()
    df["TranscriptID"] = df["TranscriptID"].str.replace("rna-", "", regex = False)
    df["NCBILink"] = "https://www.ncbi.nlm.nih.gov/gene/?term=" + df["TranscriptID"]
    df["NCBILink"] = df["NCBILink"].apply(lambda x: f'<a href="{x}" target="_blank">{x}</a>')
    df.to_html(input_dir/f"{name}.html", escape = False, index = False)
```
